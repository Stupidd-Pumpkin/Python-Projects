```python
from urllib.request import urlopen
from openpyxl import load_workbook
wb = load_workbook("data.xlsx")
import winsound
frequency = 2500  # Set Frequency To 2500 Hertz
duration = 1000  # Set Duration To 1000 ms == 1 second

#example for a mainlink "https://www.moneycontrol.com/financials/results/quarterly-results/PGH"
mainlink = "https://www.moneycontrol.com/financials/results/quarterly-results/AF30"

q = mainlink.split("/")
sheet_title = q[-1]
#sheet = wb.create_sheet(title = "Procter and Gamble Hygiene and Health Care Ltd.")
sheet = wb.create_sheet(title = sheet_title)

k = 0
for l in range(1,11):
    link = mainlink+"/"+str(l)
    f = urlopen(link)
    myfile = str(f.read())
    data = myfile.split("<td>")
    L = len(data)
    for i in range(L):
        y = data[i]
        z = y.split("<")
        z = z[0]
        z = z.replace("&#039;","")
        data[i] = z
    max=L
    for i in range(L):
        if data[i] == 'Public Share Holding':
            max = i
            break
    data = data[2:max]        
    L = len(data)
    n = 6
    if L < 38*2: break
    if L < 38*3: n = 2
    if L < 38*4: n = 3
    if L < 38*5: n = 4
    if L < 38*6: n = 5
    #print(n)    
    for i in range(int(L/n)):
        for j in range(n):
            if (j != 0):
                x = sheet.cell(row= i+1 , column = (l-1)*6+j-k+1) 
                x.value = data[i*n+j]
            elif(l == 1):
                x = sheet.cell(row= i+1 , column = (l-1)*6+j-k+1) 
                x.value = data[i*n+j]
    k = k+1;
wb.save('data.xlsx')
winsound.Beep(frequency, duration)
```
